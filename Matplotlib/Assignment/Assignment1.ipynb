```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
class_a=pd.DataFrame({
    'name':['Alice','Bob','Charlie'],
    'age':[20,21,19],
    'major':['Computer Science','Maths','Physics'],

})
```

```python
class_b=pd.DataFrame({
    'name':['David','Era'],
    'age':[22,20],
    'major':['Chemistry','Biology'],

})
```

```python
class_c=pd.DataFrame({
    'name':['Frank','Grace','Henry'],
    'age':[21,19,20],
    'major':['English','History','Art'],

})
```

```python
Combined_df=pd.concat([class_a,class_b,class_c], axis=0)
```

```python
Combined_df=Combined_df.reset_index()
Combined_df
```

```python
Combined_df['major'].unique()
```

```python
np.round(Combined_df['age'].mean(),1)
```

```python
employee=pd.DataFrame({
    'emp_id':[101,102,103,104],
    'name':['Alice','Bob','Charlie','David'],
    'dept_id':['D1','D2','D1','D3']

})
department=pd.DataFrame({
      'dept_id':['D1','D2','D3'],
      'dept_name':['Engineering','Marketing','Sales'],
      'location':['Building A','Building B','Building C']
})
```

```python
full_info=pd.merge(employee,department,how='inner',on='dept_id')
```

```python
full_info
```

```python
full_info['dept_id'].value_counts()
```

```python
full_info['dept_id'].value_counts()['D1']
```

```python
customers=pd.DataFrame({
    'cust_id':['C001','C002','C003','C004'],
    'customer_name':['john smith','john dane','Bob wilson','Alice brown'],
    'email':['john@email.com','jane@email.com','bob@email.com','alice@email.com']

})
orders=pd.DataFrame({
    'order_id':[1001,1002,1003],
    'cust_id':['C001','C003','C001'],
    'product':['Laptop','Mouse','Keyboard'],
    'amount':[1200,25,75]
})
```

```python
cust_info=pd.merge(customers,orders,how='left',on='cust_id')
cust_info
```

```python
print("Customers with no orders")
print(cust_info[cust_info['order_id'].isna()])
```

```python
print('Cutosmers with orders')
cust_info.dropna()
```

```python
multiple_orders=cust_info.groupby('cust_id').size()
print(multiple_orders[multiple_orders>1])
```
