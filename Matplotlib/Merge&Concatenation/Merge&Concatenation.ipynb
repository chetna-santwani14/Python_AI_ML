## Merging

```python
reg = pd.DataFrame(
    {
        'reg_id' : [1,2,3,4],
        'name' : ['A','B','C','D']
    }
)
```

```python
logins = pd.DataFrame(
    {
        'log_id' : [1,2,3,4],
        'name' : ['X','A','Y','B']
    }
)
```

```python
reg
```

```python
logins
```

```python
mer1=pd.merge(reg,logins,how='inner',on='name') #intersection
```

```python
mer1
```

```python
mer2=pd.merge(reg,logins,how='outer',on='name') #union
mer2
```

```python
reg
```

```python
mer3=pd.merge(reg,logins,how='left',on='name')
mer3 #left side
```

```python
logins
```

```python
mer4=pd.merge(reg,logins,how='right',on='name')
mer4
```
