```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
#Ploting using matplotlib (Functional appraoch)
plt.scatter(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter plot using matplotlib')
plt.show()
#To save the below figure
plt.savefig('Test.png')
```

```python
y=x**2
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=14)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('2D Diagram')
```

```python
#Subplots
plt.subplot(2,2,1) #The plot has 2 rows and 2 columns
plt.plot(x,y,'r--') #'r stands for red color and "--" is for dashed line '
plt.subplot(2,2,2)
plt.plot(x,y,'bo')
plt.subplot(2,2,3)
plt.plot(x,y,'g*--')
plt.subplot(2,2,4)
plt.plot(x,y,'o')
```

```python

x=np.arange(1,11)
y=3*x+6
plt.plot(x,y,'r--')
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y,'g')
plt.show()
```

```python
x=np.arange(0,5*np.pi,0.1)
ys=np.sin(x)
yc=np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,ys,'r')
plt.title('Sinx')
plt.subplots_adjust(wspace=0.5,hspace=0.5)
plt.subplot(2,1,2)
plt.plot(x,yc,'g')
plt.title('Cosx')
plt.show()
```

```python
#Bar plot
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x,y,color='y')
plt.bar(x2,y2,color='r')
plt.show()
```

# Histograms

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a,color='g') #By default, the bin size is 10
plt.xlabel('X axis')
plt.title("Histogram")
plt.show()
#Here also we can add xlabel and ylabel, but above plt.show()
```

```python
# import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 4, 9, 16, 25]

# Plot with labels
plt.plot(x, y1, label='Linear')
plt.plot(x, y2, label='Quadratic')

# Show legend
plt.legend(loc=(1.01,0.02))
plt.show()
```

```python
fig,axes=plt.subplots()
axes.plot(x,y1,marker='s',ms=20,linestyle=':',color='#634591',mew=7,mec='black',mfc='yellow',lw=3)
axes.set_title('x vs y')
plt.show() #mec:markers edge color ,mec:markers face color mew:markers edge width
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(6,4)) #Subplots can also be used like this
axes[1].plot(x,y1,'o-.',color='black')
axes[0].plot(x,y1,'s--',color='green')
```
