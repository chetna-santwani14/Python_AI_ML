```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[0,0.2,0.3,0.4]
y=[0,0.4,0.6,0.8]
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,0.5,0.5])
axes.plot(x,y)
plt.show()
```

```python
a=np.linspace(0,1,10)
```

```python
a
```

```python
b=a**2
```

```python
b
```

```python
#OOP approach
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.3,0.3,0.5,0.5])
axes1.plot(a,b)
axes2.plot(x,y)
plt.show()
```

```python
fig1=plt.figure()
fig_axes=fig1.add_axes([0.5,1,1,1.5])
a=[0.5,0,7,0.8,0.9,1]
b=[0.6,0.7,0.9,1.1,1.4,1.2]
fig_axes.plot(a,b)
plt.show()
```
