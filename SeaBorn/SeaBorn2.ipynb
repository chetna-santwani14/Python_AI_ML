```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
#sns.countplot(df['sex'])
sns.countplot(x='smoker',data=df,color='pink')
```

```python
sns.countplot(y='sex',data=df, palette=["red", "green", "blue", "purple"]) #use color :if you want single color,use palette ,if you want
#multiple colors
```

```python
sns.countplot(x='day',data=df,color='skyblue')
```

```python
#Bar Plot
sns.barplot(x='smoker',y='total_bill',hue='sex',data=df,palette=['orange','red','green'])
```

```python
#Boxplot : the lines represents percentiles 25,50 75,100 and the rings are the outliners
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(x='smoker',y='total_bill',hue='sex',data=df)
```

```python
sns.boxplot(data=df,orient='v')  #Bydefault, it takes numerical values only
```

KDE = Kernel Density Estimation.
Its a way to estimate the probability distribution of data  basically, a smooth curve that shows where data points are concentrated.

Instead of just counting data in bins (like a histogram), KDE gives you a continuous curve.

Histogram vs KDE:

Histogram  bar heights = frequency in bins.

KDE  smooth curve fitted over the data.

```python
sns.violinplot(x='day',y='total_bill',data=df)
```

Key Parts of a Violin Plot

The violin shape  indicates the distribution of the data.

Wide = more data points there.

Narrow = fewer data points.

White dot in the middle  median.

Thick black bar  interquartile range (25%75%).

Thin line (whisker)  min & max values (or clipped range).

```python
sns.violinplot(x="day", y="total_bill", hue="sex", data=df)
```

```python
sns.violinplot(x="day", y="total_bill", hue="sex", data=df, split=True)
```
