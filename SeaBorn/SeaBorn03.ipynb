```python
#Distribution Plot
plt.figure(figsize=(3,3))
sns.rugplot(data=df,x='work experience') #shows only for one feature ie work experience
```

| Feature               | Scatter Plot 🟢                             | Distribution Plot 🔵                      |
| --------------------- | ------------------------------------------- | ----------------------------------------- |
| Variables             | Two (x, y)                                  | One (only x, or only y)                   |
| Each point represents | One observation (pair of values)            | Part of frequency/density of one variable |
| Answered question     | “How does X relate to Y?”                   | “How is X spread out?”                    |
| Visualization type    | Dots on a 2D plane                          | Histogram / KDE curve                     |
| Example use           | Relationship between **salary** & **sales** | Spread of **salary**                      |


```python
plt.figure(figsize=(4,4))
sns.rugplot(df)
```

```python
#Histogram
sns.histplot(data=df,x='salary',bins=10,kde=True,color='yellow',edgecolor='pink',lw=3,ls='-')
```

```python
#KDE:KDE plots provide a smooth, continuous estimate of the probability density function
```

```python
sns.kdeplot(data=df,x='sales',fill=True,color='Pink',clip=[200000,400000])
```

```python
#CountPlot
plt.figure(figsize=(10,4))
sns.countplot(data=df,x='division',hue='training level',palette=['red','blue','yellow','green'])
```

```python
#Barplot :barplots compute a statistical estimate (default: mean) of a numerical variable for
#each category:
plt.figure(figsize=(10,4))
sns.barplot(data=df,x='division',y='salary',hue='training level',palette='Set2',estimator=np.var)
#plt.xticks(rotation=90) #rotates the x fields by 90
```
