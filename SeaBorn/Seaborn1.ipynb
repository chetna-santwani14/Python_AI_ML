Correlation with heat map

```python
corr_heat=df.select_dtypes(include=['float64', 'int64']).corr() #correlation is only possible for numerical values
display(corr_heat)
```

```python
sns.heatmap(corr_heat)
```

## Joint plot :it is used when we have 2 features

Scatterplot (or other joint plot) in the center → shows relationship between two variables.

Histograms/KDEs on the top & right margins → shows distribution of each variable individually.

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

Different values of kind:hex, reg,kde,scatter



```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
#Here ,reg stands for regression , we are also getting the regression line ,i.e best fit line ,
#and the other two are the probability density function s
```

## Displot



```python
sns.displot(df['tip'],kde=True,bins=10) #Kde stands for kernel distribution estimation
#It’s a way to estimate the probability density function (PDF) of
# a continuous variable — basically, it’s like a smooth version of a histogram.
```

```python
sns.displot(df['tip'],bins=10)
```

# Note :KDE is useful for getting a clearer picture of how your data is distributed, without relying heavily on arbitrary bin choices like in histograms.
 * In machine learning → used in probability density estimation, anomaly detection, kernel methods.

* In EDA (exploratory data analysis) → helps understand variable distributions better than raw histograms.
